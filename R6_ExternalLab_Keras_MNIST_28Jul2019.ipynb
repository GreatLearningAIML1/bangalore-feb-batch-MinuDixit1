### A MNIST-like fashion product database

In this, we classify the images into respective classes given in the dataset. We use a Neural Net and a Deep Neural Net in Keras to solve this and check the accuracy scores.

### Load tensorflow

```
import tensorflow as tf
tf.set_random_seed(42)

import numpy as np
import pandas as pd
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense,Dropout,Activation,Flatten
from keras.layers.convolutional import Conv2D,MaxPooling2D
from keras import optimizers
from keras.utils import np_utils
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```
tf.__version__
```

### Collect Data

```
(trainX, trainY), (testX, testY) = keras.datasets.fashion_mnist.load_data()
```

```
print(testY[0:5])
```

### Convert both training and testing labels into one-hot vectors.

**Hint:** check **tf.keras.utils.to_categorical()**

```
from keras.utils import to_categorical
```

```
trainX.shape
```

```
trainY.shape
```

```
#Displaying the values in different categories
L=pd.DataFrame(trainY)
L[0].value_counts()
```

```
#Represent Training & Testing samples suitable for 
#tensorflow backend
x_train=trainX.reshape(trainX.shape[0],28,28,1).astype('float32')
x_test=testX.reshape(testX.shape[0],28,28,1).astype('float32')
#for theano (1,28,28), for tensorflow (28,28,1)
```

```
x_train.shape
```

```
x_train/=255
x_test/=255
```

```
#Encoding the output class label (One-Hot Encoding)
y_train=np_utils.to_categorical(trainY,10)
y_test=np_utils.to_categorical(testY,10)
y_train[2]
```

### Visualize the data

Plot first 10 images in the triaining set and their labels.

```
import matplotlib.pyplot as plt

for x in range(0, 10):
  first_array=trainX[x]
  plt.imshow(first_array)
  plt.show()
  plt.savefig("fig.png")
  print("This image belongs to Category : ", trainY[x])
```

### Build a neural Network with a cross entropy loss function and sgd optimizer in Keras. The output layer with 10 neurons as we have 10 classes.

```
model=Sequential()
model.add(Conv2D(32,(3,3),border_mode='valid',input_shape=(28,28,1),activation='relu'))
model.add(Flatten())
model.add(Dense(10,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

### Execute the model using model.fit()

```
model.fit(x_train,y_train,batch_size=200,nb_epoch=12,verbose=1,validation_data=(x_test,y_test))
```

### In the above Neural Network model add Batch Normalization layer after the input layer and repeat the steps.

### Build the Neural Network model with 3 Dense layers with 100,100,10 neurons respectively in each layer. Use cross entropy loss function and singmoid as activation in the hidden layers and softmax as activation function in the output layer. Use sgd optimizer with learning rate 0.03.

```
model=Sequential()
model.add(Conv2D(32,(3,3),border_mode='valid',input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(100,activation='sigmoid'))
model.add(Dense(100,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='softmax'))
sgd = optimizers.SGD(lr=0.03, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

## Review model

### Run the model

```
model.fit(x_train,y_train,batch_size=200,nb_epoch=12,verbose=1,validation_data=(x_test,y_test))
```

```
score=model.evaluate(x_test,y_test)
score
```

```
y_predict = model.predict(x_test)
```
