Q6) Group teams based on their “Best position” and print the sum of their points for all positions (10 points)

Eg: Best Position     Points

         1                   25000

         2                    7000

Ans) Grouping teams as per my understanding as :

"Good Performing Team" (Been Champion or Runner-up more than 4 times)

"Average Performing Team" (Been Third or Fourth more than 4 times)

"Bad Performing Team" (Been Fifth or Sixth more than 4 times) 

Creating a new column as "Team_Performance" per the above conditions.

```python
Liga["Team_Performance"] = np.where((Liga["Champion"] >= 4) & (Liga["Runner-up"] >= 4)  , "Good Performing Team", 
                           np.where((Liga["Third"] >= 4) & (Liga["Fourth"] >= 4), "Medium Performing Team",
                           np.where((Liga["Fifth"] >= 4) & (Liga["Sixth"] >= 4),"Low Performing Team", 
                                    "Performance Not considered")))
Liga.head(10)
```

```python
# Copying the data to 3 categories
Performance_Table = Liga.loc[:,["Team","Team_Performance","Points"]]
GoodPerformance = Performance_Table.loc[Performance_Table["Team_Performance"] == "Good Performing Team"]
MediumPerformance = Performance_Table.loc[Performance_Table["Team_Performance"] == "Medium Performing Team"]
LowPerformance = Performance_Table.loc[Performance_Table["Team_Performance"] == "Low Performing Team"]
```

```python
print("Sum of Points for the Good Performing teams : ")
GoodPerformance.loc[:,["Points"]].sum()
```

```python
print("Sum of Points for the Medium Performing teams : ")
MediumPerformance.loc[:,["Points"]].sum()
```

```python
print("Sum of Points for the Low Performing teams : ")
LowPerformance.loc[:,["Points"]].sum()
```
