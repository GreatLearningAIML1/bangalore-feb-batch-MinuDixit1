Q1) Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)

```python
# importing pandas, numpy and Dataset
import pandas as pd
import numpy as np
Liga = pd.read_csv("Laliga.csv")
```

```python
Liga.head(5)
```

```python
# Making first row as index
Liga = Liga.rename(columns=Liga.iloc[0]).drop(Liga.index[0])
```

```python
# Checking if Pos is unique or not
Liga['Pos'].is_unique
```

```python
# Making Pos column as index
Liga = Liga.set_index('Pos')
```

```python
# Replacing - with 0 for better arithmetic operations
Liga = Liga.replace('-',0)
Liga.head(10)
```

Q2) Print all the teams which have started playing between 1930-1980. (5 points)

```python
#Converting Debut & Since/LastApp columns to int type by removing -## from the data
extr = Liga['Debut'].str.extract(r'^(\d{4})', expand=False)
```

```python
Liga['Debut'] = pd.to_numeric(extr)
```

```python
extr1 = Liga['Since/LastApp'].str.extract(r'^(\d{4})', expand=False)
Liga['Since/LastApp'] = pd.to_numeric(extr1)
```

```python
Liga.dtypes
```

```python
#Applying Condition
df = Liga.loc[(Liga["Debut"] >= 1930) & (Liga["Debut"] <= 1980)]
```

```python
#The teams which have started playing between 1930-1980 are : 
df["Team"]
```

Q5) Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
Liga["Winning Percent"] = (Liga["GamesWon"] / Liga["GamesPlayed"])*100
```

```python
Liga = Liga.sort_values(by=["Winning Percent"],ascending=False)
```

```python
print("The top 5 teams which have the highest Winning percentage are :")
Liga.loc[:,["Team","Winning Percent"]].head(5)
```

Q6) Group teams based on their “Best position” and print the sum of their points for all positions (10 points)

Eg: Best Position     Points

         1                   25000

         2                    7000

Ans) Grouping teams as per my understanding as :

"Good Performing Team" (Been Champion or Runner-up more than 4 times)

"Average Performing Team" (Been Third or Fourth more than 4 times)

"Bad Performing Team" (Been Fifth or Sixth more than 4 times) 

```python
print("Good Performing Team" )
grp = Liga.groupby(['Champion','Runner-up'])['Points'].sum()
grp
```

```python
print("Average Performing Team" )
grp1 = Liga.groupby(['Third','Fourth'])['Points'].sum()
grp1
```

```python
print("Bad Performing Team" )
grp2 = Liga.groupby(['Fifth','Sixth'])['Points'].sum()
grp2
```
